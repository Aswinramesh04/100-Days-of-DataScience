**Scatter Plots:**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 5, 4, 6, 2]

plt.scatter(x, y)
plt.show()
```

**Compare plots with colors:**

```python
x1 = [1, 2, 3, 4, 5]
y1 = [30, 50, 40, 60, 9]
plt.scatter(x, y, color = 'pink')

x2 = [1, 2, 3, 4, 5]
y2 = [3, 5, 4, 6, 2]
plt.scatter(x, y, color = 'green')

plt.show()
```

**Color Each Dot:**

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
x = np.random.rand(30)
y = np.random.rand(30)
colors = np.random.rand(30)

plt.scatter(x, y, c=colors, s=100, cmap='viridis', edgecolors='k')

plt.show()
```

**Pie Charts:**
the pie() function to draw pie charts:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([30, 30, 20, 20])

plt.pie(y)
plt.show()
```

```python
#labels
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Audi", "BMW", "POLO", "KIA"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
# Explode
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Audi", "BMW", "POLO", "KIA"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```
