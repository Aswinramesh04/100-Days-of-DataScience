**Seaborn:**
Seaborn is a powerful Python data visualization library that can help you create informative and attractive visualizations.

```python
import seaborn as sns

iris = sns.load_dataset("iris")
sns.histplot(data=iris)
```

**Scatter plot:**

```python
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris)
```

**Line Plots:**

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=iris)
```

**Bar Plots:**

```python
sns.lineplot(x="species", y="sepal_length", hue="species", style="species", markers=True, dashes=False, data=iris)
```

```python
sns.barplot(x="sepal_length", y="sepal_width", data=iris)
```

**Histograms:**

```python
sns.histplot(data=iris, x="sepal_length", bins=20, kde=True, color="green")
```

**Density Plots:**

```python
sns.kdeplot(x="sepal_length", data=iris)
```

```python
sns.kdeplot(data=iris, x="sepal_length", hue="species", fill=True, alpha=0.6, linewidth=1.5)
```

**Box plots** (box-and-whisker plots) provide a summary of the distribution of a numerical variable

```python
sns.boxplot(x="sepal_length", y="sepal_width", data=iris)
```

**Violin plots** combine a box plot with a kernel density estimation

```python
sns.violinplot(x="sepal_length", y="sepal_width", data=iris)
```

**Heatmaps** use color to represent the magnitude of a relationship between two categorical variables

```python
corr = iris.corr()

# Create a heatmap of the correlation between variables
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Heatmap")
plt.show()
```

**Pair Plots: **
Pair plots create scatter plots for all combinations of variables in a dataset

```python
sns.pairplot(data=iris)
```

**Joint plot** is a powerful visualization technique in seaborn that combines two different plots in one visualization

```python
sns.jointplot(data=iris)
```

**facet grids**
FacetGrid is a powerful seaborn tool that allows you to visualize the distribution of one variable as well as the relationship between two variables

```python
g = sns.FacetGrid(iris, col="species")

# Plot histogram for sepal_length in each species
g.map(sns.histplot, "sepal_length")
```
