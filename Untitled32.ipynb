```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt

# Load data from the CSV file
data = pd.read_csv('/content/100_Sales.csv', decimal='.')

# Columns to convert from string to float
string_cols = ["Region", "Country", "Item_Type", "Sales_Channel", "Order_Priority", "Ship_Date"]

# Create an OrdinalEncoder for converting string columns
encoder = OrdinalEncoder()

# Apply OrdinalEncoder to the string columns
data[string_cols] = encoder.fit_transform(data[string_cols])

# Assuming 'Total_Profit' is the target variable
X = data.drop('Total_Profit', axis=1)
y = data['Total_Profit']

# Impute missing values
imputer = SimpleImputer(strategy='mean')  # You can choose other strategies like 'median' or 'most_frequent'
X_imputed = imputer.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.3, random_state=42)

# Create and train Decision Tree Classifier
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)

# Visualize the decision tree
plt.figure(figsize=(20, 10))
plot_tree(decision_tree, filled=True, feature_names=X.columns, class_names=['0', '1'], rounded=True)
plt.show()

# Make predictions and calculate accuracy
y_pred_decision_tree = decision_tree.predict(X_test)
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
print(f"Accuracy of Decision Tree Classifier: {accuracy_decision_tree:.2f}")

# Create and train Random Forest Classifier
random_forest = RandomForestClassifier(random_state=42)
random_forest.fit(X_train, y_train)

# Make predictions and calculate accuracy
y_pred_random_forest = random_forest.predict(X_test)
accuracy_random_forest = accuracy_score(y_test, y_pred_random_forest)
print(f"Accuracy of Random Forest Classifier: {accuracy_random_forest:.2f}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Assuming 'Total_Profit' is a continuous target variable
y = data['Total_Profit']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.3, random_state=42)

# Create and train Random Forest Regressor
random_forest_regressor = RandomForestRegressor(random_state=42)
random_forest_regressor.fit(X_train, y_train)

# Make predictions and calculate mean squared error
y_pred_random_forest = random_forest_regressor.predict(X_test)
mse = mean_squared_error(y_test, y_pred_random_forest)
print(f"Mean Squared Error of Random Forest Regressor: {mse:.2f}")
```
